```python
mexicoDF = df[df["Country Name"]=="México"]
mexicoDF.columns
```

```python
# Llenar NA en columnas numéricas con la media
num_cols = mexicoDF.select_dtypes(include=['float64', 'int64']).columns
mexicoDF.loc[:, num_cols] = mexicoDF[num_cols].apply(lambda x: x.fillna(x.mean()))

# Llenar NA en columnas categóricas con la moda
cat_cols = mexicoDF.select_dtypes(include=['object', 'category']).columns
mexicoDF.loc[:, cat_cols] = mexicoDF[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
selected_indicators = ['Necesidades de anticoncepción insatisfechas (% de mujeres casadas entre 15 y 49 años de edad)',
 'Exhaustividad del registro de muertes con información sobre causa de muerte (%)',
 'Exhaustividad de registro de nacimientos (%)',
 'Exhaustividad de registro de nacimientos',
 'Completeness of birth registration',
 'Población',
 'Crecimiento de la población (%) anual',
 'Tasa de inactividad por edades',
 'Mujeres que creen que está justificado que un marido golpee a su esposa cuando descuida a los hijos (%)',
 'Mujeres que creen que está justificado que un marido golpee a su esposa cuando sale sin avisarle (%)',
 'Mujeres que creen que está justificado que un marido golpee a su esposa cuando a ella se le quema la comida (%)',
 'Mujeres que creen que está justificado que un marido golpee a su esposa cuando ella discute con él (%)',
 'Proporción de mujeres víctimas de violencia física o sexual en los últimos 12\xa0meses (% de mujeres de entre 15 y 49\xa0años)']
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Limpiando el DataFrame
# Seleccionando solo las columnas necesarias: 'Indicator Name' y los años de 1960 a 2023
years = [str(year) for year in range(1960, 2024)]
columns_needed = ['Indicator Name'] + years
mexicoDF_clean = mexicoDF[columns_needed]

# Convirtiendo el DataFrame a un formato largo para facilitar el análisis por año
mexicoDF_melted = pd.melt(mexicoDF_clean, id_vars=['Indicator Name'], 
                          value_vars=years,
                          var_name='Year', value_name='Value')

# Convertir la columna 'Value' a numérico, eliminando cualquier fila que no sea convertible
mexicoDF_melted['Value'] = pd.to_numeric(mexicoDF_melted['Value'], errors='coerce')
mexicoDF_melted = mexicoDF_melted.dropna(subset=['Value'])

# Convertir la columna 'Year' a tipo string para evitar problemas al graficar
mexicoDF_melted['Year'] = mexicoDF_melted['Year'].astype(str)

# Seleccionar un subconjunto pequeño de indicadores para graficar
indicators = mexicoDF_melted['Indicator Name'].unique()

# Graficar cada indicador respecto a los años
for indicator in selected_indicators:
    df_plot = mexicoDF_melted[mexicoDF_melted['Indicator Name'] == indicator]
    plt.figure(figsize=(10, 5))
    plt.plot(df_plot['Year'], df_plot['Value'], marker='o')
    plt.title(f"{indicator} over the Years")
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```
